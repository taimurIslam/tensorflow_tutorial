```python
import tensorflow as tf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import classification_report
%matplotlib inline
```

```python
census = pd.read_csv('census_data.csv')
```

```python
census.head()
```

```python
census['income'].unique()
```

```python
def label_fix(label):
    if label == '<=50K':
        return 0
    else:
        return 1
```

```python
census['income'] = census['income'].apply(label_fix)
```

```python
census
```

```python
x_data = census.drop('income', axis=1)
```

```python
y_levels = census['income']
```

```python
X_train, X_test, y_train, y_test = train_test_split(x_data, y_levels, test_size=0.3, random_state=101)
```

```python
census.columns
```

```python
gender = tf.feature_column.categorical_column_with_vocabulary_list("sex",
                                                                   ['Female',
                                                                    'Male'])
```

```python
occupation = tf.feature_column.categorical_column_with_hash_bucket("occupation", 
                                                                   hash_bucket_size=1000)
```

```python
marital_status = tf.feature_column.categorical_column_with_hash_bucket("marital.status", 
                                                                   hash_bucket_size=1000)

relationship = tf.feature_column.categorical_column_with_hash_bucket("relationship", 
                                                                   hash_bucket_size=1000)
education = tf.feature_column.categorical_column_with_hash_bucket("education", 
                                                                   hash_bucket_size=1000)
workclass = tf.feature_column.categorical_column_with_hash_bucket("workclass", 
                                                                   hash_bucket_size=1000)
native_country = tf.feature_column.categorical_column_with_hash_bucket("native.country", 
                                                                   hash_bucket_size=1000)
```

```python
age = tf.feature_column.numeric_column("age")

education_num = tf.feature_column.numeric_column("education.num")
capital_gain = tf.feature_column.numeric_column("capital.gain")
capital_loss = tf.feature_column.numeric_column("capital.loss")
hours_per_week = tf.feature_column.numeric_column("hours.per.week")
```

```python
feat_cols = [gender, occupation, marital_status,
             relationship, education, workclass,
             native_country, age, education_num,
             capital_gain, capital_loss, hours_per_week]
```

```python
input_func = tf.estimator.inputs.pandas_input_fn(x = X_train, y =y_train, batch_size=100, 
                                                 num_epochs=None, shuffle=True)
```

```python
model = tf.estimator.LinearClassifier(feature_columns=feat_cols)
```

```python
model.train(input_fn=input_func, steps=10000)
```

```python
pred_fn = tf.estimator.inputs.pandas_input_fn(x= X_test, batch_size=len(X_test), 
                                              shuffle=False)
```

```python
pred_gen = model.predict(input_fn=pred_fn)
```

```python
predictions = list(pred_gen)
```

```python
predictions
```

```python
final_preds = [pred['class_ids'][0] for pred in predictions]
```

```python
final_preds
```

```python
print(classification_report(y_test, final_preds))
```

```python
saver = tf.train.Saver()
saver.save(model, 'model.ckpt')
```
