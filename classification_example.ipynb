```python
import tensorflow as tf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
%matplotlib inline
```

```python
diabetes = pd.read_csv('pima-indians-diabetes.csv')
```

```python
diabetes.shape
```

```python
a = []
b = ['A','B','C','D']
for i in range(768):
    n = np.random.randint(0,4)
    a.append(b[n])
```

```python
a
```

```python
categorical_col = pd.DataFrame(data=a, columns=['Group'])
```

```python
diabetes = pd.concat([diabetes,categorical_col], axis=1)
```

```python
diabetes
```

```python
diabetes.columns
```

```python
cols_to_norm = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin',
       'BMI', 'DiabetesPedigreeFunction']
```

```python
diabetes[cols_to_norm] = diabetes[cols_to_norm].apply(lambda x: (x-x.min()) / (x.max()-x.min()))
```

```python
diabetes.head()
```

```python
num_preg = tf.feature_column.numeric_column(key = 'Pregnancies', shape=[1])
glucose = tf.feature_column.numeric_column(key = 'Glucose', shape=[1])
blood_pressure = tf.feature_column.numeric_column(key = 'BloodPressure', shape=[1])
skin_thickness = tf.feature_column.numeric_column(key = 'SkinThickness', shape=[1])
insulin = tf.feature_column.numeric_column(key = 'Insulin', shape=[1])
bmi = tf.feature_column.numeric_column(key = 'BMI', shape=[1])
diabetes_pedigree_function = tf.feature_column.numeric_column(key = 'DiabetesPedigreeFunction', shape=[1])
age = tf.feature_column.numeric_column(key = 'Age', shape=[1])
```

```python
assigned_group = tf.feature_column.categorical_column_with_vocabulary_list('Group',['A','B','C','D'])
```

```python
#assigned_group = tf.feature_column.categorical_column_with_hash_bucket('Group', hash_bucket_size=10)
```

```python
diabetes['Age'].hist(bins=20)
```

```python
age_bucket = tf.feature_column.bucketized_column(age, boundaries=[20,30,40,50,60,70,80])
```

```python
print(age_bucket)
```

```python
feat_cols = [num_preg, glucose, blood_pressure, skin_thickness, insulin, bmi, diabetes_pedigree_function,assigned_group, age_bucket]
```

```python
x_data = diabetes.drop('Outcome', axis=1)
```

```python
x_data.head()
```

```python
labels = diabetes['Outcome']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_data, labels, test_size =0.3, random_state=101)
```

```python
input_func = tf.estimator.inputs.pandas_input_fn(x = x_train, y= y_train, batch_size=10, num_epochs=1000, shuffle=True)
```

```python
model = tf.estimator.LinearClassifier(feature_columns=feat_cols, n_classes=2)
```

```python
model.train(input_fn=input_func, steps=1000)
```

```python
eval_input_func = tf.estimator.inputs.pandas_input_fn(x = x_test, y= y_test, batch_size=10, num_epochs=1000, shuffle=False)
```

```python
result = model.evaluate(eval_input_func)
```

```python
result
```

```python
pred_input = tf.estimator.inputs.pandas_input_fn(x = x_test, batch_size=10, num_epochs=1, shuffle=False) 
```

```python
predictions = model.predict(pred_input)
```

```python
dnn_model = tf.estimator.DNNClassifier(hidden_units=[10,10,10], feature_columns=feat_cols, n_classes=2)
```

```python
dnn_model.train(input_fn= input_func, steps=1000)
```

```python
embedder_group_col = tf.feature_column.embedding_column(assigned_group, dimension=4)
```

```python
feat_cols = [num_preg, glucose, blood_pressure, skin_thickness, insulin, bmi, diabetes_pedigree_function,embedder_group_col, age_bucket]
```

```python
input_func = tf.estimator.inputs.pandas_input_fn(x = x_train, y= y_train, batch_size=10, num_epochs=1000, shuffle=True)
```

```python
dnn_model = tf.estimator.DNNClassifier(hidden_units=[10,20,20,20,10], feature_columns=feat_cols, n_classes=2)
```

```python
dnn_model.train(input_fn= input_func, steps=1000)
```

```python
eval_input_func = tf.estimator.inputs.pandas_input_fn(x = x_test, y= y_test, batch_size=10, num_epochs=1, shuffle=False)
```

```python
dnn_model.evaluate(eval_input_func)
```
